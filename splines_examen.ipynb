```python
import matplotlib.pyplot as plt
import numpy as np
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) *    * 2 + d * (x - x0) ** 3
xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()
```

# Tarea
* Grafique los splines cúbicos tal que cumplan con una pendiente de $m=2$ en los puntos dados.

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) *    * 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

def Spline(x, x_i, coef):
    dx = x - x_i
    return coef["a"] + coef["b"]*dx + coef["c"]*dx**2 + coef["d"]*dx**3

def plot_tangent(x1, y1, m):
    x = np.array([x1-0.5, x1+0.5])
    y = m*(x - x1) + y1
    plt.plot(x, y, '--g', label='Tangente')

xs = [-1, 0, 1]
ys = [1, 5, 3]

s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")


plot_tangent(0, 5, 1)  # x1=0, y1=5, m=1 (pendiente en x1)

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos y tangente")
plt.legend()
plt.grid(True)
plt.show()
```
